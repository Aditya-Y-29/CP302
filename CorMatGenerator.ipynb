```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data1=pd.read_excel('Processed Data\Intrinsic Data.xlsx', sheet_name=0, usecols="B:E,H:I")
data2=pd.read_excel('Processed Data\Extrinsic Data.xlsx', sheet_name=0, usecols="B:P")
# print(data1)
data1.corr()
data2.corr()
```

```python
sns.heatmap(data1.corr(),annot=True,vmin=-1,vmax=1,center=0,cmap='vlag')
```

```python
sns.heatmap(data2.corr(),annot=True,vmin=-1,vmax=1,center=0,cmap='vlag')
```

```python
matrix=data1.corr()
mask=np.triu(np.ones_like(matrix,dtype=bool))
matrix2=data2.corr()
mask2=np.triu(np.ones_like(matrix2,dtype=bool))
```

```python
sns.heatmap(data1.corr(),annot=True,vmin=-1,vmax=1,center=0,cmap='vlag',mask=mask)
```

```python
sns.heatmap(data2.corr(),annot=True,vmin=-1,vmax=1,center=0,cmap='vlag',mask=mask2)
```
